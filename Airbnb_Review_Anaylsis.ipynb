```python
import pandas as pd
```

```python
# Load the data

df = pd.read_csv('joined_reviews_and_listings.csv')
```

```python
df
```

```python
df.info()
```

```python
# Question 1:

What property and room types are most common among listings, and how do they correlate with review scores and guest sentiment?
```

```python
# Find the most common property and room types
property_counts = df['property_type'].value_counts()
room_counts = df['room_type'].value_counts()
```

```python
# Display the most common property and room types
print("Most common property types:")
print(property_counts.head())

print("\nMost common room types:")
print(room_counts.head())
```

```python
# Group by property type and room type

correlation_analysis = df.groupby(['property_type', 'room_type'])[
    ['review_scores_cleanliness', 'review_scores_value', 'positive_sentiment']
].mean()
```

```python
# Display the correlation analysis results

print("\nAverage review scores and sentiment by property and room types:")
print(correlation_analysis.sort_values(by='positive_sentiment', ascending=False).head())
```
